## Data cleaning
### Column names

```python
autos.columns
```

```python
columns = ['dateCrawled', 'name', 'seller', 'offerType', 'price', 'abtest',
       'vehicleType', 'registration_year', 'gearbox', 'powerPS', 'model',
       'kilometer', 'registration_month', 'fuelType', 'brand',
       'unrepaired_damage', 'ad_created', 'nrOfPictures', 'postalCode',
       'lastSeen']

columns_new = []

for col in columns:
    col = col.lower()
    columns_new.append(col)
```

```python
autos.columns = columns_new
autos.head(5)
```

In the code above we modified all variables names to lowercase, also manually modifying some of them to depict the data they contain more accurately ('unrepaired_damage', for example). For that task, we first stored them in a separate list, and when modifications were done, we reintroduced them in the dataframe. 

### Initial explorations
We'll now briefly dig into the data to identify what other data cleaning tasks we need, looking for text columns with almost all values being equal or numeric ones with their information stored as text, among others.

```python
autos.describe(include='all')
```

There are some columns with almost all values being the same, such as 'seller', 'offertype', 'gearbox'... So they appear to be candidates for deletion if they don't offer important information. Aside from this, other columns like 'vehicletype' and, specially, 'unrepaired_damage', have missing rows, so we need to take a look at them to see if we can impute or delete them too. 

```python
print(autos.dtypes)
```

Looking again at the columns' types, there are some of type 'object' that should be 'date' instead, but apparently there are no mismatches between text and numeric data.

```python
autos["datecrawled"] = pd.to_datetime(autos["datecrawled"], format="%Y-%m-%d %H:%M:%S")
autos["lastseen"] = pd.to_datetime(autos["lastseen"], format="%Y-%m-%d %H:%M:%S")
autos["ad_created"] = pd.to_datetime(autos["ad_created"], format="%Y-%m-%d %H:%M:%S")
```

```python
print(autos.dtypes)
autos.head(5)
```

Let's now have a look at those columns with almost all identical values:

```python
print(autos["abtest"].value_counts())
print(autos["offertype"].value_counts())
print(autos["gearbox"].value_counts())
print(autos["fueltype"].value_counts())
print(autos["unrepaired_damage"].value_counts())
print(autos["seller"].value_counts())
```

Watching this, we could tell that, although values are incorrectly coded, all columns may offer some degree of information except for 'offertype' and 'seller', so it may be a good idea to discard them, so we can free space and add clarity. 

```python
autos = autos.drop(["offertype", "seller"], axis = 1)
```

```python
# Configurar la opción para mostrar números sin notación científica
pd.set_option('display.float_format', lambda x: '%.3f' % x)

print(autos["price"].describe())
print(autos["kilometer"].describe())
```

```python
autos = autos.loc[autos["price"].between(100, 1000000)]
```

```python
print(autos["price"].describe())
```

We've successfully discarded outliers from the price column, which clearly had incorrect entries that could lead to mistaken analyses. Let's now get back to the date columns, as our objective is to also remove possible errors.

```python
print(autos["ad_created"].min())
print(autos["ad_created"].max())
print(autos["lastseen"].min())
print(autos["lastseen"].max())
print(autos["datecrawled"].max())
print(autos["datecrawled"].min())
```

They all seem to be correctly coded. What about 'registration_year'?

```python
print(autos["registration_year"].describe())
print(autos["registration_year"].value_counts())
print(autos["registration_month"].describe())
print(autos["registration_month"].value_counts())
```

```python
autos = autos.loc[autos["registration_year"].between(1950, 2017)]
```

### Aggregations

Now that we have semi-clean data, we can perform some initial tasks. For example, let's look at the average price on the top-5 offered brands on eBay:

```python
top_brands = autos["brand"].value_counts().head(5).index
print(top_brands)
```

```python
agg = {}
for brand in top_brands:
    filtered = autos.loc[autos["brand"] == brand]
    mean = filtered["price"].mean()
    agg[brand] = round(mean, 2)
```

```python
agg
```

What about the least offered? Are there differences?

```python
bottom_brands = autos["brand"].value_counts().tail(5).index
print(bottom_brands)
```

```python
agg_bot = {}
for brand in bottom_brands:
    filtered = autos.loc[autos["brand"] == brand]
    mean = filtered["price"].mean()
    agg_bot[brand] = round(mean, 2)

agg_bot
```

Having a quick look, the least offered cars are not necessarily more expensive. Maybe it has to do with the cars being old instead? Let's look at the average again, but from another column:

```python
age = {}
for brand in top_brands:
    filtered = autos.loc[autos["brand"] == brand]
    mean = filtered["registration_year"].mean()
    age[brand] = mean
```

```python
age
```

```python
age_bot = {}
for brand in bottom_brands:
    filtered = autos.loc[autos["brand"] == brand]
    mean = filtered["registration_year"].mean()
    age_bot[brand] = mean
```

```python
age_bot
```

As we can see, there are some notable differences in the mean age of the top and bottom brands. Let's store all info from the top firms together!

```python
price = pd.Series(agg)
year = pd.Series(age)
df = pd.DataFrame(price, columns=["mean_price"])
df["mean_year"] = round(year, 0)
df
```

Although making a mean from yearly data is not the best idea, it may gives us a general approach to the car's status, and how well it fits into the current driveways: maybe there are environmental restrictions that don't allow citizens to drive cars above a certain age threshold in their cities!
